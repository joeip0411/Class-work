# Check unigram

```python
f = open('sample_100uni.txt','r')
l_s = f.read().split('\n')
l_s = sorted(l_s)[1:]

g = open('uni100.txt','r')
l_o = g.read().split('\n')
l_o = sorted(l_o)[1:]
```

```python
count_total = 0
total = 0

for i in range(len(l_s)):
    sample = eval(l_s[i][11:])
    output = eval(l_o[i][11:])
    
    # transform to dict
    
    d_s = {}
    d_o = {}
    
    for i in sample:
        d_s[i[0]] = i[1]
    
    for j in output:
        d_o[j[0]] = j[1]
        
    count = 0
    
    for k in d_s.keys():
        if d_s[k] <= d_o.get(k,-100) +5 and d_s[k] >= d_o.get(k,-100) -5:
            count +=1
            count_total += 1
        total +=1
    
    print(count, len(d_s))

print(count_total/total)
```
