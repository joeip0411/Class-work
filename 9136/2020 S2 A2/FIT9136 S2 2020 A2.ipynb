# 2. Zipf's Law: Common Statistics

## 2.1 Find all the unique word in the cleaned file. Save all the clean words along with their frequecies, and sort in descending order, based on word's frequency in file vocab.txt.

```python
f = open('clean.txt')
s = f.read()
s = s.lower()
s = s.split()

count_dict = {}

for word in s:
    if count_dict.get(word) is None:
        count_dict[word] = 1
    else:
        count_dict[word] += 1
```

```python
count_dict
```

```python
count_list = []

for k,v in count_dict.items():
    count_list.append((k , v))
```

```python
count_list
```

```python
sort_list = sorted(count_list, key = lambda tup: tup[1], reverse = True)
```

```python
sort_list
```

```python
f = open('vocab.txt', 'w')

index = 0

for i in sort_list:
    print(str(index)+'. '+ i[0]+':'+str(i[1]), file = f)
    index += 1

f.close()
```

```python
def find_suggestion(df):
    string = input('Enter your stmt: ')
    
    last_word = string.split()[-1]
    
    try:
        suggestion = df[last_word].idxmax()
    except:
        return ('Cannot give a suggestion')
        
    return 'Suggestyed word = ' +suggestion
```
